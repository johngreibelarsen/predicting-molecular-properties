#### Merging data sources into one training data set and one test data set 

Code for joining data is from this kernel: https://www.kaggle.com/seriousran/just-speed-up-calculate-distance-from-benchmark

```python
def map_atom_info(df, atom_idx):
    df = pd.merge(df, structures, how = 'left',
                  left_on  = ['molecule_name', f'atom_index_{atom_idx}'],
                  right_on = ['molecule_name',  'atom_index'])
    
    df = df.drop('atom_index', axis=1)
    df = df.rename(columns={'atom': f'atom_{atom_idx}',
                            'x': f'x_{atom_idx}',
                            'y': f'y_{atom_idx}',
                            'z': f'z_{atom_idx}'})
    return df
```

```python
train = map_atom_info(train, 0)
train = map_atom_info(train, 1)
```

```python
train.head(10)
```

```python
test = map_atom_info(test, 0)
test = map_atom_info(test, 1)
```

### Feature engineering
#### Distances between atoms

```python
train_vec_0 = train[['x_0', 'y_0', 'z_0']].values
train_vec_1 = train[['x_1', 'y_1', 'z_1']].values
test_vec_0 = test[['x_0', 'y_0', 'z_0']].values
test_vec_1 = test[['x_1', 'y_1', 'z_1']].values
```

```python
import numpy as np

train['dist'] = np.linalg.norm(train_vec_0 - train_vec_1, axis=1)
test['dist'] = np.linalg.norm(test_vec_0 - test_vec_1, axis=1)
```

```python
train.head()
```

#### No of atomic bond distances

```python
train['atomic_bond_distances'] = train['type'].apply(lambda x: int(x[0]))
test['atomic_bond_distances'] = train['type'].apply(lambda x: int(x[0]))
```

```python
train.head()
```

#### Make a copy of data for first training

```python
columns_to_use_v1 = [
'atom_index_0',
'atom_index_1',    
'atom_0',
'atom_1',
'dist',
'atomic_bond_distances'
]
```

For the above columns I will consider atom_index's ordinal data together with no_of_atomic_bond_distances. I will argue that atom's are pure labels, that is, nominal data.

LabelEncoder could be used for the ordinal data but as they are already ordered integers we will leave them be. One Hot Encoding is used for nominal data.

```python
train_X = train[columns_to_use_v1].copy()
train_y = train['scalar_coupling_constant']
test_X = test[columns_to_use_v1].copy()
```

```python
atom_0_OHE = pd.get_dummies(train_X['atom_0'], prefix='atom_0')
atom_1_OHE = pd.get_dummies(train_X['atom_1'], prefix='atom_1')

train_X = pd.concat([train_X, atom_0_OHE], axis=1)
train_X = pd.concat([train_X, atom_1_OHE], axis=1)

train_X = train_X.drop(['atom_0', 'atom_1'], axis=1)


atom_0_OHE_test = pd.get_dummies(test_X['atom_0'], prefix='atom_0')
atom_1_OHE_test = pd.get_dummies(test_X['atom_1'], prefix='atom_1')

test_X = pd.concat([test_X, atom_0_OHE_test], axis=1)
test_X = pd.concat([test_X, atom_1_OHE_test], axis=1)

test_X = test_X.drop(['atom_0', 'atom_1'], axis=1)
```

#### Data splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(train_X, train_y, random_state = 13, test_size = 0.25)
```

### The model

```python
from catboost import CatBoostRegressor
```

```python
categorical_features_indices = np.where(train_X.dtypes != np.float)[0]
print(categorical_features_indices)
```

```python
basic_model = CatBoostRegressor(task_type = "GPU", devices='0', gpu_ram_part=0.95, gpu_cat_features_storage='GpuRam',  iterations=3000, random_seed=42, loss_function='MAE', custom_metric=['MAE'], eval_metric='MAE', grow_policy='SymmetricTree', border_count=255, random_strength=2, learning_rate=0.03, l2_leaf_reg=6, depth=3)
# Old model: bestTest = 3.697536446, bestIteration = 11999 using: iterations=12000, random_seed=42, loss_function='MAE', custom_metric=['MAE'], eval_metric='MAE', grow_policy='SymmetricTree', border_count=255, random_strength=2, learning_rate=0.03, l2_leaf_reg=6, depth=3
# This model: bestTest ~ 2.7 bestIteration = 11999 using: iterations=12000, random_seed=42, loss_function='MAE', custom_metric=['MAE'], eval_metric='MAE', grow_policy='SymmetricTree', border_count=255, random_strength=2, learning_rate=0.03, l2_leaf_reg=6, depth=3
# This model: bestTest = 2.53 bestIteration = 19999 using: iterations=20000, random_seed=42, loss_function='MAE', custom_metric=['MAE'], eval_metric='MAE', grow_policy='SymmetricTree', border_count=255, random_strength=2, learning_rate=0.03, l2_leaf_reg=6, depth=3
```

```python
basic_model.fit(X_train, y_train, 
          cat_features=categorical_features_indices, 
          eval_set=(X_valid, y_valid),
          logging_level='Verbose', plot=True)
```

```python
# Recreate the type column

def find_atom_symbol(H, C, N):
    if H == 1: 
        return 'H'
    elif C == 1:
        return 'C'
    elif N == 1:
        return 'N'
    return ''

X_valid['type'] = 'None'

X_valid['type'] = X_valid.apply(lambda row: str(row['atomic_bond_distances']) + 'H' + find_atom_symbol(row['atom_1_H'], row['atom_1_C'], row['atom_1_N']), axis=1)
```

```python
X_valid.head(10)
```

```python
# Competition metric from here https://www.kaggle.com/uberkinder/efficient-metric

def group_mean_log_mae(y_true, y_pred, groups, floor=1e-9):
    maes = (y_true-y_pred).abs().groupby(groups).mean()
    return np.log(maes.map(lambda x: max(x, floor))).mean()
```

```python
score = group_mean_log_mae(y_valid, y_pred, X_valid['type'])
print(f'score {score}')
# 1000  iter 1.52793
# 3000  iter 1.25545
# 20000 iter 0.82817
```
